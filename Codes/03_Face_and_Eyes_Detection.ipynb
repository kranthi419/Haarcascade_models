### Import Libraries

```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

### Create Classifiers

```python
face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_classifier = cv2.CascadeClassifier('haarcascade_eye.xml')
```

### Open Image

```python
image = cv2.imread('eye_face.jpg')
```

### Fix Image

```python
fix_img = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

### Face Classifier

```python
faces = face_classifier.detectMultiScale(image,1.3,5)
```

### If No Face Detected

```python
if faces is ():
    print('NO FACES FOUND')
```

### It's a Kind of Magic

```python
def detect_face(fix_img):
    face_rects = face_classifier.detectMultiScale(fix_img)
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(fix_img,
                     (x,y),
                     (x+w,y+h),
                     (255,0,0),10)
    
    eye_rects = eye_classifier.detectMultiScale(fix_img)
    
    for (x,y,w,h) in eye_rects:
        cv2.rectangle(fix_img,
                     (x,y),
                     (x+w,y+h),
                     (255,255,255),10)
    return fix_img
```

```python
result = detect_face(fix_img)
plt.imshow(result)
```
